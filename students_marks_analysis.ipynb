```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("std_data.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Drop unnamed column

```python
df= df.drop("Unnamed: 0",axis=1)
print(df.head())
```

# change weekly study hours
#replace function used

```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("05-Oct","5-10")
```

```python
df.head()
```

# Gender distribution

```python
plt.figure(figsize=(3,2))
sns.countplot(data = df , x="Gender")
plt.show()
```

# for the exact count of female and the male in numbers show in the graph

```python
plt.figure(figsize=(3,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.show
```

#From the above chart we have analyzed that the count of female students are more than the male students

```python
#we are finding that the does parents education affects on the students marks
```

```python
gb= df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
sns.heatmap(gb, annot = True)
plt.show()
```

```python
#from the above chart we have conclude that the parents education does affect on their marks
```

```python
gb1= df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
sns.heatmap(gb1 , annot=True)
plt.show()
```

```python
#From above heatmap Parents marital status have no imapact or neglegible impact on students scores
```
